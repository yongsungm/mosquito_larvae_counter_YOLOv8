```python
from ultralytics import YOLO
```

```python
# load a pretrained model (recommended for training)
model = YOLO("yolov8n.pt")

# Tune hyperparameter for 30 epochs
model.tune(data='data.yaml', epochs=30, iterations=300, optimizer='AdamW', plots=False, save=False, val=False)
```

```python
# load tuned hyperparameters and reset model for training
hyperparameters = 'runs/detect/tune/best_hyperparameters.yaml'
model = YOLO("yolov8n.pt")
```

```python
#train the model
results = model.train(data="data.yaml", epochs=500, cfg=hyperparameters)
```

```python
model_path = "runs/detect/train/weights/best.pt"
model = YOLO(model_path)
```

```python
# Test the model
metrics = model.val(data='test.yaml')  # run evaluation on test data
metrics.box.map    # map50-95
metrics.box.map50  # map50
metrics.box.map75  # map75
metrics.box.maps   # a list contains map50-95 of each category
```
